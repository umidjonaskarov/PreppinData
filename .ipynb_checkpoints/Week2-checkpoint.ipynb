```python
# importing libraries 

import pandas as pd
import csv
import re
import numpy as np
from datetime import datetime as dt
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importing inputs

df1 = pd.read_csv('Data/Week2/Transactions.csv')
df2 = pd.read_csv('Data/Week2/Swift codes.csv')

#merging two dataframes 

df = pd.merge(df1, df2)
df[0:1000]
```

```python
# removing dashes
df['Sort Code'] = df['Sort Code'].str.replace('-','')
df
```

```python
#adding country code 

df['Country code'] = "GB"
df
```

```python
def generate_iban(df):
    country_code = df['Country code']
    check_digits = df['Check Digits']
    swift_code = df['SWIFT code']
    account_number = df['Account Number']
    
    # Combine the components to form the IBAN
    iban = f"{country_code}{check_digits}{swift_code}{account_number}"
    
    return iban

df['IBAN'] = df.apply(generate_iban, axis=1)

df 
```

```python
# Outputs #plotlib

transactions_by_bank = df.groupby("Bank")["Transaction ID"].count()
transactions_by_bank

plt.pie(transactions_by_bank, labels=transactions_by_bank.index, autopct='%1.1f%%')
plt.title("Number of Transactions by Bank ")
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.show()
```
