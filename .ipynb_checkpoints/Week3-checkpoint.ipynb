```python
# importing libraries 

import pandas as pd
import csv
import re
import numpy as np
from datetime import datetime as dt
import matplotlib as plt
import locale

```

```python
# importing input.csv

df1 = pd.read_csv("data/Week3/Input.csv")
df2 = pd.read_csv("data/Week3/Targets.csv")
```

```python
#extracting bank code into new column 

def extract_bank_code(transaction_code):
    bank_code = re.search(r'([A-Z]+)-', transaction_code)
    if bank_code:
        return bank_code.group(1)
    else:
        return None

df1['Bank'] = df1['Transaction Code'].apply(extract_bank_code)
df1.query('Bank == "DSB"')
```

```python
# changing column data into Online or In-Person 


df1['Online or In-Person'] = df1['Online or In-Person'].apply(change_name)

def change_name(value):
    if value == 1:
        return "Online"
    else:
        return "In-Person"

df1
```

```python
df1 = df1.query('Bank == "DSB"')
df1
```

```python
# Convert 'Transaction Date' column to datetime
df1['Transaction Date'] = pd.to_datetime(df1['Transaction Date'], format='%d/%m/%Y %H:%M:%S')

# Extract the quarter and assign it to a new column called 'Quarter'
df1['Quarter'] = df1['Transaction Date'].dt.quarter
df1
```

```python
by_quarter = df1.groupby(["Quarter", 'Online or In-Person'])["Value"].sum()
by_quarter
```
