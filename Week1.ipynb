```python
# importing libraries 

import pandas as pd
import csv
import re
import numpy as np
from datetime import datetime as dt
import matplotlib as plt
```

```python
# importing input.csv

df = pd.read_csv("PreppinData/Data/Week1/input.csv")
```

```python
# extracting bank name 

def extract_bank_code(transaction_code):
    bank_code = re.search(r'([A-Z]+)-', transaction_code)
    if bank_code:
        return bank_code.group(1)
    else:
        return None

df['Bank code'] = df['Transaction Code'].apply(extract_bank_code)
```

```python
# rearranging table columns

cols = ['Transaction Code', 'Bank code','Value', 'Customer Code', 'Online or In-Person', 'Transaction Date']
df = df[cols]
df
```

```python
# replacing 1,2 into online or in-person
df['Online or In-Person'] = np.where(df['Online or In-Person'] == 1, 'Online', 'In-Person')
```

```python
df
```

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], format='%d/%m/%Y %H:%M:%S') #transforming into date format 
df
```

```python
df['Weekday'] = df['Transaction Date'].dt.day_name() #applying day name function into transformed date
df
```

```python
#renaming columns 

df = df.rename(columns={"Bank code": "Bank", "Weekday": "Transaction Day"})
df
```

```python
# output 1

transactions_by_bank = df.groupby("Bank")["Value"].sum()
transactions_by_bank
```

```python
# output 2 

transactions_by_bank_week = df.groupby(["Bank", "Transaction Day", 'Online or In-Person'])["Value"].sum()
transactions_by_bank_week
```

```python
# output 3

bank_customer_code = df.groupby(['Customer Code', 'Bank'])['Value',].count()
bank_customer_code
```
